```python
import pandas as pd
from rdkit.Chem import AllChem
import numpy as np
import xgboost as xgb
```

```python
file_train = "../../dataset/Training_set_HEPoP_after_filtering.json"
file_test = "../../dataset/Holdout_test_set_HEPoP_after_filtering.json"
train_df = pd.read_json(file_train)
test_df = pd.read_json(file_test)
```

```python
# test_df[["reactant_smiles", "product_smiles"]]


def get_fingerprints_from_df(df):
    failed = 0
    labels, reaction_fps = [], []
    reactant_smiles, product_smiles = [], []
    functional_groups_reacted = []
    for ind, row in df.iterrows():
        reactant_smi = row["reactant_smiles"]
        product_smi = row["product_smiles"]
        
        label = row["dG_sp"]
        array_temp = np.zeros((1, 1024))
        try:
            if type(reactant_smi) == str:
                reactant_fps = [
                    AllChem.GetMorganFingerprintAsBitVect(
                        AllChem.MolFromSmiles(reactant_smi), 2, nBits=1024
                    )
                ]
            else:
                reactant_fps = [
                    AllChem.GetMorganFingerprintAsBitVect(
                        AllChem.MolFromSmiles(i), 2, nBits=1024
                    )
                    for i in reactant_smi
                ]
            if type(product_smi) == str:
                product_fps = [
                    AllChem.GetMorganFingerprintAsBitVect(
                        AllChem.MolFromSmiles(product_smi), 2, nBits=1024
                    )
                ]
            else:
                product_fps = [
                    AllChem.GetMorganFingerprintAsBitVect(
                        AllChem.MolFromSmiles(i), 2, nBits=1024
                    )
                    for i in product_smi
                ]

            for i in reactant_fps:
                array_temp += i
            for i in product_fps:
                array_temp += i
            reaction_fps.append(array_temp.reshape(-1))
            labels.append(label)
            reactant_smiles.append(reactant_smi)
            product_smiles.append(product_smi)
            functional_groups_reacted.append(row["functional_group_reacted"])
        except:
            failed += 1
    reactant_fps = np.array(reaction_fps)
    labels = np.array(labels)
    print("failed: ", failed)
    return reactant_fps, labels, reactant_smiles, product_smiles, functional_groups_reacted


reactant_fps_train, labels_train, _, _, _ = get_fingerprints_from_df(train_df)
reactant_fps_test, labels_test, reaction_smiles_test, product_smiles_test, functional_groups_test = get_fingerprints_from_df(test_df)
```

```python
# scale the labels
mean, std = np.mean(labels_train), np.std(labels_train)
labels_train = (labels_train - mean) / std
labels_test = (labels_test - mean) / std

# Use "hist" for training the model.
reg = xgb.XGBRegressor(tree_method="hist", device="cuda")

reg.fit(reactant_fps_train, labels_train)
```

```python
print(reg.score(reactant_fps_train, labels_train))
print(reg.score(reactant_fps_test, labels_test))
```

```python
# get performance metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error

y_pred = reg.predict(reactant_fps_test)
print("MSE: ", mean_squared_error(labels_test * std, y_pred * std))
print("R2: ", r2_score(labels_test * std, y_pred * std))
print("MAE: ", mean_absolute_error(labels_test * std, y_pred * std))
```

```python
from skopt.searchcv import BayesSearchCV
from skopt.space import Real, Integer


xgb_temp = xgb.XGBRegressor()
reg = BayesSearchCV(
    xgb_temp,
    {
        "colsample_bytree": Real(0.5, 0.99),
        "max_depth": Integer(5, 25),
        "lambda": Real(0, 0.25),
        "learning_rate": Real(0.1, 0.25),
        "alpha": Real(0, 0.2),
        "eta": Real(0, 0.1),
        "gamma": Real(0, 0.1),
        "n_estimators": Integer(50, 1000),
        "objective": ["reg:squarederror"],
        "tree_method": ["gpu_hist"],
    },
    n_iter=20,
    verbose=4,
    cv=3,
)
reg.fit(reactant_fps_train, labels_train)
```

```python
y_pred = reg.predict(reactant_fps_test)
print("MSE: ", mean_squared_error(labels_test * std, y_pred * std))
print("R2: ", r2_score(labels_test * std, y_pred * std))
print("MAE: ", mean_absolute_error(labels_test * std, y_pred * std))

# best
# MSE:  28.764100574436675
# R2:  0.8449550918513228
# MAE:  3.4711014961723756

# new best
#MSE:  26.077014136345195
#R2:  0.8594390861936214
#MAE:  3.08899534194241
```

```python
preds_unscaled = y_pred * std + mean
labels_unscaled = labels_test * std + mean
print(len(preds_unscaled), len(labels_unscaled), len(reaction_smiles_test), len(product_smiles_test), len(functional_groups_test))
# make a df 
df = pd.DataFrame({"preds": preds_unscaled, "labels": labels_unscaled, "reaction_smiles": reaction_smiles_test, "product_smiles": product_smiles_test, "functional_groups": functional_groups_test})
df.to_csv("xgb_test_preds.csv", index=False)
```
