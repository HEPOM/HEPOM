```python
# Make plot of dataset size vs. MAE error
import matplotlib.pyplot as plt
import scienceplots


dataset_size_1 = [100, 1000, 10000, 12000]
dataset_size_2 = [114, 1100, 5000, 6954]

mae_error_1 = [0.74, 0.57, 0.36, 0.33]
mae_error_2 = [0.61, 0.35, 0.28, 0.27]

with plt.style.context(['science', 'notebook', 'ieee']):
    plt.figure()
    # change figure size
    plt.figure(figsize=(7, 5))
    plt.plot(dataset_size_1, mae_error_1)
    plt.plot(dataset_size_2, mae_error_2)
    # log scale on x axis
    plt.xscale('log')
    # make publication format
    plt.xlabel('Dataset size')
    plt.ylabel('MAE error')
    plt.title('MAE error vs. dataset size - QTAIM Features')
    plt.legend(['w/out QTAIM', 'w/QTAIM'])
    # plt.tight_layout()
    #plt.show()
    # save figure as svg
    plt.savefig('dataset_size_vs_mae_error.png', format='png', dpi=1200)
```
